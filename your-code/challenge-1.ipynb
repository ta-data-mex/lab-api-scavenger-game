```python
# import the required libraries
import requests
import pandas as pd
```

```python
# use the request method to obtain general information.
data_madrid = requests.get('https://api.github.com/repos/ironhack-datalabs/mad-oct-2018')
data_madrid = data_madrid.json()
data_madrid
```

```python
# Obtain with the request method the fork information
data_madrid_fork = requests.get('https://api.github.com/repos/ironhack-datalabs/mad-oct-2018/forks')
data_madrid_fork = data_madrid_fork.json()
data_madrid_fork
```

```python
# Obtain easily the languages. 

data_madrid_lang = requests.get('https://api.github.com/repos/ironhack-datalabs/mad-oct-2018/languages')
data_madrid_lang = data_madrid_lang.json()
data_madrid_lang
```

```python
# Obtain the languages of the 'languages'
new_data_madrid_fork = pd.DataFrame(data_madrid_fork)
list(set(new_data_madrid_fork['language']))
```
