```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
git_h = requests.get("https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018/git/refs")
```

```python
df = pd.DataFrame(git_h.json())
df.head()
```

```python
objetos = json_normalize(df['object'])
```

```python
shas= list(objetos['sha'])
```

```python
print(type(shas))
```

```python
commitslst = []
for sha in shas:
    commits = requests.get(f"https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018/git/commits/{sha}")
    commits.json()
    commitslst.append(commits.json())
```

```python
commitslst
```

```python
normalizado = json_normalize(commitslst)
```

```python
fechas = normalizado['committer.date']
```
